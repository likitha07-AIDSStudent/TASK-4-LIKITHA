```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_excel("cleaned_Ecommerce_Dataset.xlsx")
```

```python
df.head()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
product_counts=df["Product"].value_counts()
product_counts
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x="Product", data=df)
plt.title("Most Ordered Products")
plt.xlabel("Product")
plt.ylabel("Number Of Orders")
plt.xticks(rotation=45)
plt.show()
```

Insight: This bar chart shows the number of orders for each product. Products with taller bars are ordered more frequently, indicating higher customer demand.

```python
payment_counts = df["PaymentMethod"].value_counts()

payment_counts
```

```python
plt.figure(figsize=(7,7))

payment_counts.plot(
    kind="pie",
    autopct="%1.1f%%",
    startangle=90
)

plt.title("Payment Method Distribution")
plt.ylabel("")

plt.show()
```

Insight: This pie chart shows the percentage of orders placed using each payment method. The largest slice represents the payment method preferred by most customers.

```python
order_status = df["OrderStatus"].value_counts()

order_status
```

```python
plt.figure(figsize=(7,7))

order_status.plot(
    kind="pie",
    autopct="%1.1f%%",
    startangle=90
)

plt.title("Order Status Distribution")
plt.ylabel("")

plt.show()
```

Insight:
This pie chart shows the distribution of order statuses. It helps us understand how many orders are delivered, shipped, pending, or cancelled. A higher percentage of delivered orders indicates successful order fulfillment.

```python
referral_counts = df["ReferralSource"].value_counts()

referral_counts
```

```python
plt.figure(figsize=(8,5))

sns.countplot(x="ReferralSource", data=df)

plt.title("Referral Source Analysis")
plt.xlabel("Referral Source")
plt.ylabel("Number of Customers")
plt.xticks(rotation=45)

plt.show()
```

Insight:
This bar chart shows the number of customers coming from different referral sources. The referral source with the tallest bar is the most effective channel for attracting customers.

```python
plt.figure(figsize=(8,5))

plt.hist(df["TotalPrice"], bins=20)

plt.title("Distribution of Total Price")
plt.xlabel("Total Price")
plt.ylabel("Number of Orders")

plt.show()
```

### Observation
- This histogram shows how the Total Price values are distributed.
- Most orders fall within a specific price range.
- It helps identify whether there are many low-value or high-value orders.

```python
plt.figure(figsize=(8,5))

plt.scatter(df["Quantity"], df["TotalPrice"])

plt.title("Quantity vs Total Price")
plt.xlabel("Quantity")
plt.ylabel("Total Price")

plt.show()
```

### Observation
- The scatter plot shows the relationship between Quantity and Total Price.
- As the quantity increases, the total price generally increases.
- This indicates a positive relationship between these two variables.

```python
plt.figure(figsize=(8,5))

sns.boxplot(y=df["TotalPrice"])

plt.title("Box Plot of Total Price")
plt.ylabel("Total Price")

plt.show()
```

### Observation
- The box plot shows the distribution of Total Price values.
- It helps identify outliers (orders with unusually high or low total prices).
- Most order values are concentrated within the main range shown by the box.

```python
df["Date"] = pd.to_datetime(df["Date"])
```

```python
daily_sales = df.groupby("Date")["TotalPrice"].sum()
```

```python
plt.figure(figsize=(10,5))

daily_sales.plot()

plt.title("Sales Trend Over Time")
plt.xlabel("Date")
plt.ylabel("Total Sales")

plt.show()
```

### Observation
- The line chart shows how total sales change over time.
- Peaks represent days with higher sales, while dips indicate lower sales.
- This trend can help identify busy and slow sales periods.
